## Filtrado con Loc y Iloc

loc

```python
df_books.loc[:4, ['Name','Author']] # Este si incluye el stop
```

```python
df_books.loc[:, ['Reviews']] * -1
```

```python
df_books.loc[:, ['Author']] == "JJ Smith"
```

iloc

```python
df_books.iloc[:] #iloc trabaja con los indices de las columnas
```

```python
df_books.iloc[:,0:3]
```

```python
df_books.iloc[1,3] * -1
```
