## Combinando dataframes


### Merge y concat


```python
df1 = pd.DataFrame({"A": ["a0","a1","a2","a3"],
         "B": ["b0","b1","b2","b3"],
         "C": ["c0","c1","c2","c3"],
         "D": ["d0","d1","d2","d3"],})

df2 = pd.DataFrame({"A": ["a4","a5","a6","a7"],
         "B": ["b4","b5","b6","b7"],
         "C": ["c4","c5","c6","c7"],
         "D": ["d4","d5","d6","d7"],})
```

```python
pd.concat([df1,df2],ignore_index=True) 
# ignore_index = True reordena los indices segun la entrada
```

```python
pd.concat([df1,df2],axis=1) 
```

Merge

```python
izq = pd.DataFrame({"key": ["k0","k1","k2","k3"],
    "A": ["a0","a1","a2","a3"],
    "B": ["b0","b1","b2","b3"]})
der = pd.DataFrame({"key": ["k0","k1","k2","k3"],
    "C": ["c0","c1","c2","c3"],
    "D": ["d0","d1","d2","d3"]})
```

```python
izq.merge(der , on="key") 
# on = n Hace el merge con una llave en comun, parecido al join de SQL
```

```python
izq = pd.DataFrame({"key": ["k0","k1","k2","k3"],
    "A": ["a0","a1","a2","a3"],
    "B": ["b0","b1","b2","b3"]})
der = pd.DataFrame({"key2": ["k0","k1","k2","k3"],
    "C": ["c0","c1","c2","c3"],
    "D": ["d0","d1","d2","d3"]})
```

```python
izq.merge(der , left_on="key", right_on="key2")
# en caso de que las que hagamos merge no coincidan en el nombre
```

```python
izq = pd.DataFrame({"key": ["k0","k1","k2","k3"],
    "A": ["a0","a1","a2","a3"],
    "B": ["b0","b1","b2","b3"]})
der = pd.DataFrame({"key2": ["k0","k1","k2",np.nan],
    "C": ["c0","c1","c2","c3"],
    "D": ["d0","d1","d2","d3"]})
```

```python
izq.merge(der , left_on="key", right_on="key2", how="right")
# how = "right", "left", "inner" , "outer"
# para hacer los distintos tipos de joins
```

# Join

Join trabaja con los indices en vez de las columnas  como hace el merge
index match


```python
izq= pd.DataFrame({"A": ["a0","a1","a2"],
                 "B": ["b0","b1","b2"]},
                 index=["k0",'k1',"k2"])
der= pd.DataFrame({"C": ["c0","c1","c2"],
                   "D": ["d0","d1","d2"]},
                 index=["k0",'k2',"k3"])
```

```python
izq.join(der, how="inner") 
```

```python
izq.join(der, how="outer") 
```

```python
def two_times(value):
    return value*2
```
